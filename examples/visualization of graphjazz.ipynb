```python
from graphjazz.plot import twcircle
import matplotlib.pyplot as plt
```

```python
bytep=[0,1,1,0,1,0,0,1,0,1,0,1]
```

```python
fig=plt.figure()
ax = plt.axes()
twcircle.set_twcircle(ax)
twcircle.ntonic(ax,bytep,rcircle=0.1)
plt.show()
```

```python
# You can make custom ntonic visualization:
```

```python
fig=plt.figure()
ax = plt.axes()
twcircle.set_twcircle(ax)
twcircle.ntonic(ax,bytep,rcircle=0.2,linecolor="gray",oncolor="orange",offcolor="black")
plt.show()
```
